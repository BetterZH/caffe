```python
%matplotlib inline

import os
import pylab
from pylab import *
from collections import OrderedDict
###from skimage import io, transform
import skimage
import io
import networkx
from collections import deque
#import pydot

from IPython import display   #needed to render in notebook
#from IPython.display import display, SVG

plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# Make sure that caffe is on the python path:
caffe_root = '../'  # this file is expected to be in {caffe_root}/notebooks
import sys
sys.path.insert(0, caffe_root + 'python')
import caffe
from caffe import imagenet


# Some function definitions
def figsize(width,height):
    rcParams['figure.figsize'] = (width,height)
        
def norm01(arr):
    arr = arr.copy()
    arr -= arr.min()
    arr /= arr.max()
    return arr

def showimage(im, c01=False, skipcolorflip=False):
    if c01:
        # switch order from c,0,1 -> 0,1,c
        im = im.transpose((1,2,0))
    if im.ndim == 3 and not skipcolorflip:
        # Change from BGR -> RGB
        im = im[:, :, ::-1]
    plt.imshow(im)
    #axis('tight')

def showimagesc(im, c01=False, skipcolorflip=False):
    showimage(norm01(im), c01, skipcolorflip)

def tile_images(data, padsize=1, padval=0, c01=False, width=None):
    '''take an array of shape (n, height, width) or (n, height, width, channels)
    and visualize each (height, width) thing in a grid. If width = None, produce
    a square image of size approx. sqrt(n) by sqrt(n), else calculate height.'''
    data = data.copy()
    if c01:
        # Convert c01 -> 01c
        data = data.transpose(0, 2, 3, 1)
    data -= data.min()
    data /= data.max()
    
    # force the number of filters to be square
    if width == None:
        width = int(np.ceil(np.sqrt(data.shape[0])))
        height = width
    else:
        assert isinstance(width, int)
        height = int(np.ceil(float(data.shape[0]) / width))
    padding = ((0, width*height - data.shape[0]), (0, padsize), (0, padsize)) + ((0, 0),) * (data.ndim - 3)
    data = np.pad(data, padding, mode='constant', constant_values=(padval, padval))
    
    # tile the filters into an image
    data = data.reshape((height, width) + data.shape[1:]).transpose((0, 2, 1, 3) + tuple(range(4, data.ndim + 1)))
    data = data.reshape((height * data.shape[1], width * data.shape[3]) + data.shape[4:])
    data = data[0:-padsize, 0:-padsize]  # remove excess padding
    
    return data

def vis_square(data, padsize=1, padval=0, c01=False):
    data = tile_images(data, padsize, padval, c01)
    showimage(data, c01=False)
```

```python
# NetworkX Graph

def compute_pos(graph, nodes_fixed=None, kk=None):
    #rng = np.random.RandomState(0)
    #pos0 = dict([(node, rng.uniform(0,1, 2)) for node in graph.nodes()])
    ##print pos0
    #pos=networkx.spring_layout(graph, pos=pos0) # positions for all nodes

    pos = networkx.graphviz_layout(graph,prog='twopi',args='')
    return pos
    #pos = networkx.graphviz_layout(graph, prog='twopi', root='n00001740')

    #pos=networkx.spring_layout(graph, pos=pos) # positions for all nodes

    #pos=networkx.spectral_layout(graph)
    
    # Method N+1: circle
    posarr = array([pos[key] for key in sorted(pos.keys())])
    posarr = (posarr-(posarr.min(0)+posarr.max(0))/2.0)
    radius = sqrt((posarr**2).sum(1).max())
    posarr = (posarr.T * (radius / (sqrt((posarr**2).sum(1)) + 1e-10))).T
    posarr -= posarr.min(0)  # Put into [0,1] box for spring
    posarr /= posarr.max(0)
    for ii,key in enumerate(sorted(pos.keys())):
        pos[key] = tuple(posarr[ii,:])
    pos=networkx.spring_layout(graph, pos=pos, fixed = nodes_fixed, k=kk)

    return pos


def plot_graph(graph, pos, ax = None, node_labels = None, node_colors = None, node_size = 50):
    '''Plots a graph directly'''

    if ax == None:
        ax = gca()
    
    if isinstance(node_size, dict):
        # Create list instead
        node_size = [node_size[node] for node in graph.nodes()]

    networkx.draw(graph, pos=pos, ax=ax, with_labels=False,
                  node_size = node_size,
                  node_color=[node_colors[node] for node in graph.nodes()],
                  )
    # Prune labels to the nodes actually in the graph. draw_networkx_labels isn't
    # happy otherwise (since pos is limited to the graph)
    if node_labels != None:
        node_labels = dict([(key, node_labels[key]) for key in graph.nodes()])
        networkx.draw_networkx_labels(graph, pos, node_labels)


def graph_to_svg(graph, pos, node_labels = None, node_colors = None, node_size = 50):
    """return the SVG of a matplotlib figure generated from a graph"""
    #print 'node_labels are', node_labels
    #print 'node_colors are', node_colors

    fig = figure()
    #ax = fig.add_subplot(111)
    ax = gca()
        
    plot_graph(graph, pos, ax = ax, node_labels = node_labels,
               node_colors = node_colors, node_size = node_size)
    
    output = io.BytesIO()
    fig.savefig(output, format='svg')
    close(fig)
    return output.getvalue()
```

### Create Graph

```python
def construct_graph(leaves, is_a, Nclasses = 1000, Ndeep = 30):
    graph = networkx.DiGraph()
    
    node_labels = {}
    node_colors = {}
    node_edges_through = {}
    cmap = matplotlib.pyplot.cm.jet
    for ii in range(Nclasses):
        synset = leaves[ii]
        worklist = [[0, synset]]
        count = 0
        while len(worklist) > 0 and count < Ndeep:
            depth,synset = worklist.pop()
            graph.add_node(synset)
            #node_labels[synset] = synset
            node_labels[synset] = ''
            node_colors.setdefault(synset, (1,0,0))
            node_edges_through[synset] = node_edges_through.get(synset, 0) + 1
            generics = is_a.get(synset, [])
            #if len(generics) > 1:
            #    print 'for synset', synset, labels[synset]
            #    for gen in generics:
            #        print '  generic:', gen, labels[gen]
            for g_syn in generics:
                #node_labels[g_syn] = g_syn
                node_labels[g_syn] = ''
                node_colors.setdefault(g_syn, (1,1,1))
                node_edges_through[g_syn] = node_edges_through.get(g_syn, 0) + 1
                graph.add_node(g_syn)
                graph.add_edge(synset, g_syn)
                
                worklist.append([depth+1, g_syn])
            count += 1
    return graph

def color_graph(graph, leaves, cmap):
    '''Creates a dict mapping node -> color'''
    colors = {}
    for node in graph.nodes():
        colors[node] = (1,1,1,1)
    Nleaves = len(leaves)
    for ii in range(Nleaves):
        synset = leaves[ii]
        colors[synset] = cmap(float(ii)/Nleaves)
    return colors
```

Prune some edges

```python
def prune_to_single_path(graph, leaves, root = 'n00001740'):
    n_predecessors = dict([(node, len(graph.predecessors(node))) for node in graph.nodes()])
    for edge in graph.edges():
        src,dst = edge
        graph.edge[src][dst]['weight'] = 1.0/n_predecessors[dst]
        
    paths = networkx.single_source_dijkstra_path(graph.reverse(copy=True), root)
    
    # Mark all edges that are on a shortest path
    edge_on_path = dict([(edge, False) for edge in graph.edges()])
    for ii in range(Nclasses):
        leaf = leaves[ii]
        for jj in range(len(paths[leaf])-1):
            #print 'edge is', paths[leaf][jj+1], 'to', paths[leaf][jj]
            src,dst = paths[leaf][jj+1],paths[leaf][jj]
            edge_on_path[(src,dst)] = True
    
    edges_to_keep = set([key for key in edge_on_path.keys() if edge_on_path[key]])
    nodes_to_keep = set([node for edge in edges_to_keep for node in edge])
    
    min_graph = networkx.DiGraph(graph)
    print 'Before: %d nodes, %d edges' % (min_graph.number_of_nodes(), min_graph.number_of_edges())
    min_graph.remove_edges_from(set(min_graph.edges()) - edges_to_keep)
    min_graph.remove_nodes_from(set(min_graph.nodes()) - nodes_to_keep)
    print 'After: %d nodes, %d edges' % (min_graph.number_of_nodes(), min_graph.number_of_edges())
    return min_graph
```

Compute leaves under given higher-level categories

```python
def leaves_under(graph, leaves):
    '''For each node in graph, compute the number of unique
    leaf nodes feeding into it and return the results as a
    dict.'''
    memo = dict()
    for node in graph.nodes():
        memoized_leaves_under(memo, node, graph, leaves)
    return memo

def memoized_leaves_under(memo, node, graph, leaves):
    if not node in memo:
        if node in leaves:
            memo[node] = set((node,))
        else:
            leaves_below_this_node = set()
            for pred in graph.predecessors(node):
                leaves_below_this_node.update(memoized_leaves_under(memo, pred, graph, leaves))
            memo[node] = leaves_below_this_node

    return memo[node]
```
