```python
results_dir = '../examples/imagenet/results/140403_060842_9aa0083_priv_init0/'
model_def_file = '../examples/imagenet/imagenet_deploy.prototxt'
model0 = '../examples/imagenet/results/140403_060842_9aa0083_priv_init0/caffe_imagenet_train_iter_0'
model1 = '../examples/imagenet/results/140403_060847_9aa0083_priv_init1/caffe_imagenet_train_iter_0'

net0 = caffe.Net(model_def_file, model0)
net0.set_phase_test()
net0.set_mode_cpu()
net1 = caffe.Net(model_def_file, model1)
net1.set_phase_test()
net1.set_mode_cpu()
```
